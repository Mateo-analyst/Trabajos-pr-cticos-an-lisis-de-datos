```python
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
```

```python
estudiantes = pd.read_csv("/content/drive/MyDrive/CSV/Copia de Recolección de datos personales análisis de datos explorador (respuestas) - Respuestas de formulario 1.csv")
estudiantes['Estatura'] = estudiantes['Estatura'].astype(str).str.extract(r'(\d+(?:[,.]\d+)?)')estudiantes
```

```python
estudiantes.head()
```

```python
estudiantes.dtypes
```

```python
estudiantes['Estatura:'] = estudiantes['Estatura:'].astype(str).str.extract(r'(\d+(?:[,.]\d+)?)')
estudiantes['Estatura:'] = estudiantes['Estatura:'].str.replace(',', '.', regex=False)
estudiantes['Estatura:'] = estudiantes['Estatura:'].astype(float)
estudiantes['Estatura:'] = np.where(estudiantes['Estatura:'] > 10, estudiantes['Estatura:'] / 100, estudiantes['Estatura:'])
estudiantes['Estatura:'] = estudiantes['Estatura:'].replace(180, 18)
estudiantes
```

```python
estudiantes['Edad:'] = estudiantes['Edad:'].astype(str).str.extract(r'(\d+(?:[,.]\d+)?)')
estudiantes['Edad:'] = estudiantes['Edad:'].astype(float)
estudiantes['Edad:'] = estudiantes['Edad:'].astype(int)
estudiantes
```

```python
estudiantes['Peso:'] = estudiantes['Peso:'].astype(str).str.extract(r'(\d+(?:[,.]\d+)?)')
estudiantes['Peso:'] = estudiantes['Peso:'].astype(float)
estudiantes['Peso:'] = estudiantes['Peso:'].fillna(estudiantes['Peso:'].mean())
estudiantes['Peso:'] = estudiantes['Peso:'].astype(int)
estudiantes
```

```python
estudiantes[estudiantes.select_dtypes('object').columns] = estudiantes.select_dtypes('object').apply(lambda x: x.astype(str).str.title())

estudiantes
```

```python
estudiantes['Sexo:'] = estudiantes['Sexo:'].str.lower()
Diccionario_sexo = estudiantes['Sexo:'].unique()
Diccionario_sexo

Diccionario_sexo = {
    'hombre 	': 'Masculino',
    'maculino ': 'Masculino',
    'masculino': 'Masculino',
    'm': 'Masculino',
    'mujer': 'Femenino',
    'femenino': 'Femenino',
    'f': 'Femenino'
}

estudiantes['Sexo:'] = estudiantes['Sexo:'].str.lower().replace(Diccionario_sexo)
estudiantes['Sexo:'] = estudiantes['Sexo:'].str.title()
estudiantes
```

```python
estudiantes['Mano dominante:'] = estudiantes['Mano dominante:'].str.lower()
Diccionario_mano = estudiantes['Mano dominante:'].unique()
Diccionario_mano

Diccionario_mano = {
    'izquierda' : 'izquierda',
    'derecha' : 'derecha',
    'zurda' : 'izquierda',
    'r': 'derecha',
    'mano cambiada' : 'ambidiestro'
}

estudiantes['Mano dominante:'] = estudiantes['Mano dominante:'].str.lower().replace(Diccionario_mano)
estudiantes['Mano dominante:'] = estudiantes['Mano dominante:'].str.title()
estudiantes
```

```python
estudiantes['EPS:'] = estudiantes['EPS:'].str.lower()
Diccionario_eps = estudiantes['EPS:'].unique()
Diccionario_eps

Diccionario_eps = {
    'sura' : 'sura',
    'sumimedical' : 'sumimedical',
    'colsanitas' : 'colsanitas',
    'nueva eps' : 'nueva eps',
    'salud total' : 'salud total',
    'ecopetrol' : 'ecopetrol',
    'sura ' : 'sura',
    'sanitas' : 'sanitas',
    'zura' : 'sura',
    'sumimedical ' : 'sumimedical',
    'savia salud' : 'savia salud'
}

estudiantes['EPS:'] = estudiantes['EPS:'].str.lower().replace(Diccionario_eps)
estudiantes['EPS:'] = estudiantes['EPS:'].str.title()
estudiantes
```

```python
estudiantes['Talla de zapatos:'] = estudiantes['Talla de zapatos:'].replace({'Cuarenta Y Uno': 41})
estudiantes
```

```python
diccionario_egresados = estudiantes['Universidad egresado:'].unique()
diccionario_egresados

diccionario_egresados = {
    'No' : 'No egresado',
    'Nan' : 'No egresado',
    'Sena' : 'Sena',
    'Udea' : 'UdeA',
    'Upb' : 'UPB',
    'Universidad Pontificia Bolivariana' : 'UPB',
    'Luiz Hamigó' : 'Luiz amigó',
    'Ninguna ' : 'No egresado'
}

estudiantes['Universidad egresado:'] = estudiantes['Universidad egresado:'].replace(diccionario_egresados)
estudiantes
```
