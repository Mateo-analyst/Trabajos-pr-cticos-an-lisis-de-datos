```python
#Cambiando el tipo de dato fecha de object a datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])
df.info()
```

```python
#A las fehchas faltantes les hice un cambio por el promedio de fechas ya que solo era un 5% de los datos aprox
df['Fecha'] = df['Fecha'].fillna(df['Fecha'].mode().iloc[0])
df['Fecha'].isna().sum().sum()
```

```python
#Sacar categoria valores para el mapeo y cambio de categorias
df['Categoria'].unique()
```

```python
#Sacar de producto valores para el mapeo y cambio de categorias
df['Producto'].unique()
```

```python
#darle categoria a cada producto para que los que tian nan cambien a su categoria correspondiente
mapa_categorias = {
    'Micrófono' : 'Accesorios',
    'Webcam' : 'Periféricos',
    'Impresora' : 'Oficina',
    'SSD' : 'Almacenamiento',
    'Tablet' : 'Electrónica',
    'USB' : 'Almacenamiento',
    'Auriculares' : 'Accesorios',
    'Monitor' : 'Periféricos',
    'HDD' : 'Almacenamiento',
    'Teclado' : 'Periféricos',
    'Router' : 'Conectividad',
    'Smartwatch' : 'Accesorios',
    'Laptop' : 'Electrónica',
    'Smartphone' : 'Smartphone',
    'Mouse' : 'Periféricos'
}

df['Categoria'] = df['Producto'].map(mapa_categorias)

print(df)
```

```python
#Precios unitarios faltantes
#Valores unico de precios unitarios segun el producto

# Agrupar por producto y obtener precios unitarios únicos
precios_unicos = df.groupby('Producto')['Precio_Unitario'].unique()

print(precios_unicos)
```

```python
#se organizan y se pone valores a los precios unitarios, solo si tienen unidades vendidas y Total_venta

#(df['Unidades'].notna()) & (df['Valor_Total'].notna()) → asegura que ambos valores existan.
#(df['Unidades'] != 0) → evita dividir por cero.
#df['Valor_Total'] / df['Unidades'] → calcula el precio unitario solo si se cumple la condición.
#np.nan → deja los casos imposibles como NaN.
#fillna(...) → solo rellena los NaN existentes, sin alterar precios ya correctos.
import numpy as np

df['Precio_Unitario'] = np.where(
        (df['Precio_Unitario'].isna()) &
        (df['Cantidad'].notna()) & (df['Total_Venta'].notna()) & (df['Cantidad'] != 0),  # condición segura
        df['Total_Venta'] / df['Cantidad'],  # calcular precio unitario
        df['Precio_Unitario']                               # si falta alguno o unidades=0, dejar NaN
    )
df
```

```python
df['Precio_Unitario'].isna().sum().sum()
```

```python
#Mirar el porcentaje de datos faltantes en la columna Precio Unitario
#y se corrobora que sea el mismo porcentaje en Total_Venta
porcentaje_nulos_PrecioUnitario = (df['Precio_Unitario'].isna().sum().sum() / len(df)) * 100
print(porcentaje_nulos_PrecioUnitario)

porcentaje_nulos_TotalVenta = (df['Total_Venta'].isna().sum().sum() / len(df)) * 100
print(porcentaje_nulos_TotalVenta)
```

```python
#Ahora voy a borrar todas las filsa que tengan valor nan en su columna Precio_unitario
df = df.dropna(subset=['Precio_Unitario'])
df
```

```python
#cambio los nan de la columna metodo de pago por la moda ya que solo es el 8.1% del total
df['Metodo_Pago'] = df['Metodo_Pago'].fillna(df['Metodo_Pago'].mode().iloc[0])
df['Metodo_Pago'].isna().sum().sum()
```

```python
#cambio los nan de la columna Ciudad_Cliente por la moda ya que solo es el 11.6% del total
df['Ciudad_Cliente'] = df['Ciudad_Cliente'].fillna(df['Ciudad_Cliente'].mode().iloc[0])
df['Ciudad_Cliente'].isna().sum().sum()
```
